```python
import json
```

# Load data from JSON FILE

```python
def load_data(filename):
    with open(filename, "r") as f:
        data = json.load(f)
    return data
```

```python
data = load_data("store_data.json")
print(data, type(data))
```

# Clean and structure the data

```python
def clean_data(data):
    str_to_num = {"one":1, "two":2, "three":3, "four":4, "five":5}
    unique_set = set()
    cleaned_data = []
    for user in data:
        row_rating = str(user["rating"]).strip().lower()
        if(row_rating in str_to_num):
            row_rating = str_to_num[row_rating]
        user["rating"] = row_rating

        row_age = user.get("age")
        if(row_age == None):
            user["age"] = None

        if(user["name"] in unique_set):
            continue
        unique_set.add(user["name"])
        cleaned_data.append(user)
    return cleaned_data
```

# dump data into tha JSON FILE

```python
def dump_data(data):
    with open("store_data.json", "w") as f:
        json.dump(data, f)
```

```python
dump_data(data)
data = clean_data(data)
```

# finding meaningful insights form the cleaned data

```python
def meaningful_insights(data):
    # Average Rating...
    all_ratings = 0    
    for user in data:       
        all_ratings += float(user["rating"])
    print("Average ratings = ",all_ratings/len(data))

    # percentage of poor ratings
    total_ratings = 0
    poor_ratings = 0
    for user in data:
        if(float(user["rating"]) != None):
            total_ratings += 1
        if((float(user["rating"]) != None) and (float(user["rating"]) < 3)):
            poor_ratings += 1
        ppr = poor_ratings *100/total_ratings
    print("percentage of poor ratings <3 = ",ppr)
```

```python
meaningful_insights(data)
```

# Build Recommendation Features

```python
def recomm(data):
    recommendation = []
    for user in data:
        curr_recomm = {}
        curr_recomm["name"] = user["name"]

        if(float(user["rating"]) >= 4):
            curr_recomm["Brand"] = "Apple"
        else:
            curr_recomm["Brand"] = "Samsung"
        recommendation.append(curr_recomm)
    return recommendation
```

```python
recomm(data)
```
